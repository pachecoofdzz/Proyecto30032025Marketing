```python
# Hago todas las importaciones necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargo los datos leyendo el csv y uso on_bad_lines='skip' para saltar las líneas con errores o problemas
df = pd.read_csv('../data/archivoMarketingLimpio.csv')
```

```python
frecuencia_canales = df['type'].value_counts()
roi_promedio_por_canal = df.groupby('type')['roi'].mean()
canal_mas_frecuente = frecuencia_canales.idxmax()
canal_mejor_roi = roi_promedio_por_canal.idxmax()
```

```python
print(f"Canal más frecuente: {canal_mas_frecuente}")
print(f"Canal con mejor ROI: {canal_mejor_roi}")
```

```python
# Gráfico de barras para la frecuencia de los canales
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='type', order=frecuencia_canales.index, palette='Set2')
plt.title('Frecuencia de Uso de los Canales de Marketing')
plt.xlabel('Canal de Marketing')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Diagrama de caja para la distribución de ROI por canal
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='type', y='roi', palette='Set2')
plt.title('Distribución del ROI por Canal de Marketing')
plt.xlabel('Canal de Marketing')
plt.ylabel('ROI')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
'''
¿Qué tipo de campaña genera más ingresos en promedio y cuál tiene mejor conversión?
Objetivo: Compara los diferentes tipos de campaña para identificar cuáles generan más ingresos y tienen mejor tasa de conversión.

Enfoque sugerido: Analiza el rendimiento financiero (ingresos, beneficio neto) y las métricas de efectividad (tasa de conversión, ROI) por tipo de campaña. Busca si existen relaciones entre estas métricas que puedan explicar el éxito de ciertos tipos.

Visualizaciones recomendadas: Gráfico de barras para ingresos promedio y tasa de conversión por tipo, diagrama de dispersión para correlación entre estas métricas
'''
# Paso 1: Análisis de la Distribución del ROI

# Histograma del ROI
plt.figure(figsize=(10, 6))
sns.histplot(df['roi'], bins=30, kde=True, color='skyblue')
plt.title('Distribución del ROI')
plt.xlabel('ROI')
plt.ylabel('Frecuencia')
plt.show()
```

```python
# Boxplot del ROI
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='roi', palette='Set2')
plt.title('Distribución del ROI')
plt.xlabel('ROI')
plt.show()
```

```python
# Filtramos solo las columnas numéricas
numeric_columns = df.select_dtypes(include=[np.number])

# Calculamos la matriz de correlación solo para las columnas numéricas
correlation_matrix = numeric_columns.corr()

# Visualizamos la matriz de correlación
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Heatmap de Correlaciones entre variables numéricas')
plt.show()
```

```python
# Gráfico de dispersión ROI vs. Presupuesto
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='budget', y='roi', hue='type', palette='Set2')
plt.title('Relación entre Presupuesto y ROI')
plt.xlabel('Presupuesto')
plt.ylabel('ROI')
plt.show()
```

```python
# Gráfico de dispersión ROI vs. Tipo de campaña (si hay una columna 'campaign_type')
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='type', y='roi', palette='Set2')
plt.title('Distribución del ROI por Tipo de Campaña')
plt.xlabel('Tipo de Campaña')
plt.ylabel('ROI')
plt.show()
```

```python
# Boxplot para ver la distribución del ROI y detectar outliers
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['roi'], color='lightgreen')
plt.title('Distribución del ROI - Boxplot')
plt.xlabel('ROI')
plt.show()
```

```python
# Paso 1: Boxplot comparativo de la tasa de conversión entre B2B y B2C
plt.figure(figsize=(10, 6))
sns.boxplot(x='target_audience', y='conversion_rate', data=df, palette='Set1')
plt.title('Comparación de la Tasa de Conversión entre Audiencias B2B y B2C')
plt.xlabel('Audiencia')
plt.ylabel('Tasa de Conversión')
plt.show()
```

```python
# Paso 2: Análisis comparativo por canal
plt.figure(figsize=(12, 6))
sns.barplot(x='target_audience', y='conversion_rate', hue='channel', data=df, ci=None)
plt.title('Tasa de Conversión por Audiencia y Canal')
plt.xlabel('Audiencia')
plt.ylabel('Tasa de Conversión')
plt.show()
```

```python
# Paso 1: Ordenar las campañas por beneficio neto y seleccionar las top 5-10
top_campaigns = df.sort_values(by='eficienciaCosto', ascending=False).head(10)
```

```python
# Gráfico de barras para analizar la distribución de canales en las top 10 campañas
plt.figure(figsize=(10, 6))
sns.countplot(x='channel', data=top_campaigns, palette='Set1')
plt.title('Distribución de Canales en las Top 10 Campañas Más Rentables')
plt.xlabel('Canal')
plt.ylabel('Número de Campañas')
plt.show()
```

```python
# Gráfico de barras para analizar la distribución de tipos de campaña en las top 10 campañas
plt.figure(figsize=(10, 6))
sns.countplot(x='type', data=top_campaigns, palette='Set2')
plt.title('Distribución de Tipos de Campaña en las Top 10 Más Rentables')
plt.xlabel('Tipo de Campaña')
plt.ylabel('Número de Campañas')
plt.show()
```

```python
# Filtrar las campañas que tienen un ROI mayor a 0.5 y un ingreso mayor a 500,000
high_performance_campaigns = df[(df['roi'] > 0.5) & (df['revenue'] > 500000)]
```

```python
# Mostrar las campañas de alto rendimiento
print("Campañas de alto rendimiento (ROI > 0.5 y ingresos > 500,000):")
high_performance_campaigns
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(high_performance_campaigns[['campaign_name', 'roi', 'revenue']].set_index('campaign_name').T, annot=True, cmap='YlGnBu', fmt='.2f')
plt.title('Campañas de Alto Rendimiento (ROI > 0.5 y Ingresos > 500,000)')
plt.show()
```

```python
# Crear gráfico de dispersión con cuadrantes
plt.figure(figsize=(10, 6))
sns.scatterplot(data=high_performance_campaigns, x='roi', y='revenue', hue='channel', palette='viridis', s=100, alpha=0.7)

# Añadir líneas de corte para dividir los cuadrantes
plt.axhline(y=500000, color='red', linestyle='--', label='Ingresos = 500,000')
plt.axvline(x=0.5, color='blue', linestyle='--', label='ROI = 0.5')

plt.title('Campañas de Alto Rendimiento (ROI > 0.5 y Ingresos > 500,000)')
plt.xlabel('ROI')
plt.ylabel('Ingresos')
plt.legend()
plt.show()
```

```python
# Agrupar por mes y tipo de campaña, y calcular el promedio de ROI
monthly_campaign_roi = df.groupby(['start_date', 'type'])['roi'].mean().unstack()

# Crear un heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(monthly_campaign_roi, annot=True, cmap='YlGnBu', fmt='.2f', linewidths=0.5)
plt.title('Rendimiento mensual por tipo de campaña')
plt.xlabel('Tipo de campaña')
plt.ylabel('Mes')
plt.show()
```
